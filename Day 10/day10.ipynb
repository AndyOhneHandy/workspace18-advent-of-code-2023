```python
#########################
# Advent of Code 2023   #
# Day 10                #
#########################

import math
file = open("input_10_23.txt","r")
text = file.read()
data = text.split("\n")
data = [list(x) for x in data if x]
```

```python
# Part 1

def create_empty_matrix(data):
    matrix = []
    for row,pipes in enumerate(data):
        matrix += [list(len(pipes)*".")]
    return matrix

def check_empty(row, column, dist_matrix):
    return dist_matrix[row][column] == '.'

def update_matrix(row, column, dist_matrix, count):
    dist_matrix[row][column] = count
    return dist_matrix

def find_next(coords_list, data, dist_matrix):
    row = coords_list[-1][0]
    column = coords_list[-1][1]

    sign = data[row][column]
    dist_matrix = update_matrix(row, column, dist_matrix, len(coords_list)-2)

    if sign == "|":
        a1 = [row-1, column]
        a2 = [row+1, column]
        if coords_list[-2] == a1:
            coords_list += [a2]
        else:
            coords_list += [a1]

    if sign == "-":
        a1 = [row, column-1]
        a2 = [row, column+1]
        if coords_list[-2] == a1:
            coords_list += [a2]
        else:
            coords_list += [a1]

    if sign == "L":
        a1 = [row-1, column]
        a2 = [row, column+1]
        if coords_list[-2] == a1:
            coords_list += [a2]
        else:
            coords_list += [a1]

    if sign == "J":
        a1 = [row-1, column]
        a2 = [row, column-1]
        if coords_list[-2] == a1:
            coords_list += [a2]
        else:
            coords_list += [a1]
 
    if sign == "7":
        a1 = [row, column-1]
        a2 = [row+1, column]
        if coords_list[-2] == a1:
            coords_list += [a2]
        else:
            coords_list += [a1]
    
    if sign == "F":
        a1 = [row, column+1]
        a2 = [row+1, column]
        if coords_list[-2] == a1:
            coords_list += [a2]
        else:
            coords_list += [a1]

    next_coords = coords_list[-1]

    if check_empty(next_coords[0],next_coords[1], dist_matrix) == False:
        return None, dist_matrix
    
    return coords_list, dist_matrix

def find_start(data):
    start_pos = []
    for row, pipes in enumerate(data):
        for column, pipe in enumerate(pipes):
            if pipe == "S":
                start_pos += [row, column]
    # start is an F
    return start_pos

start = find_start(data)
start_coord_list = [['not','integer'],start]
# start is an F
data[start_coord_list[1][0]][start_coord_list[1][1]] = 'F'
dist_matrix = create_empty_matrix(data)

while start_coord_list != None:
    start_coord_list, dist_matrix = find_next(start_coord_list,data,dist_matrix)

sr = start[0]
sc = start[1]
print(f"Part 1: {math.ceil(max(dist_matrix[sr-1][sc],dist_matrix[sr+1][sc],dist_matrix[sr][sc-1],dist_matrix[sr][sc+1],)/2)}")
```
