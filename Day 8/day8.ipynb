```python
#########################
# Advent of Code 2023   #
# Day 8                 #
#########################

file = open("input_8_23.txt","r")
text = file.read()
data = text.split("\n")
data = [x for x in data if x]
instrucions = list("LRLRRRLRRRLLLRLRRLLRLRRRLRLRRRLRLRRRLRLRRRLRRRLRLLRRRLRLRLRRLRRLRLRRLRRLRRLLRRRLRRRLRRLRRLRRLRRRLLRRLRLRRLRLRRLRRLRLRRLRRLLRLRRRLRRLRRRLLRLRLRLLRLLRLLRLRRLLRRLRLRLRRLRLLRRRLLRRRLRRLLRRRLRRRLRLRRRLLRRRLRLRRRLLLRRRLRLRLRRRLRRRLRRRLRLRRLLLRRLRRRLLRLRRRLRLRLLLRRLRLRRRLRLRRRR")
```

```python
data = [''.join(filter(str.isalpha, node)) for node in data]

direction_dict = {}
for node in data:
    listofnode = list(node)
    direction_dict["".join(listofnode[0:3])] = ["".join(listofnode[3:6]),"".join(listofnode[6:9])]

def crawl_dict(instrucions, location = 'AAA'):
    counter = 0
    while location != 'ZZZ':
        for dir in instrucions:
            if dir == "L":
                location = direction_dict[location][0]
                counter += 1
            if dir == "R":
                location = direction_dict[location][1]
                counter += 1
            if dir == 'ZZZ':
                break
    return counter

print(f"Part 1: {crawl_dict(instrucions)}")
```

```python
starting_nodes = []

for a in direction_dict:
    if list(a)[2] == "A":
        starting_nodes += [a]

def move(node, dir):
    if dir == "L":
        dir = 0
    if dir == "R":
        dir = 1
    return direction_dict[node][dir]

def check_end(list_nodes):
    for node in list_nodes:
        if list(node)[2] == 'Z':
            continue
        else:
            return False
    return True

def crawl_dict2(intructions, starting_nodes = starting_nodes):
    counter = 0
    while not check_end(starting_nodes):
        for dir in instrucions:
            counter += 1
            if dir == "L":
                starting_nodes = [direction_dict[location][0] for location in starting_nodes]
            if dir == "R":
                starting_nodes = [direction_dict[location][1] for location in starting_nodes]
            if dir == check_end(starting_nodes):
                break
            if counter % 100000000 == 0:
                print(counter)
    return counter     

print(f"Part 2: {crawl_dict2(instrucions)}")   
    
```
