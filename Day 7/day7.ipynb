```python
#########################
# Advent of Code 2023   #
# Day 7                 #
#########################

file = open("input_7_23.txt","r")
text = file.read()
data = text.split("\n")
data = [x.split(" ") for x in data if x]
```

```python
def recode_hand(hand):
    hand = list(hand)
    values = {
        "T": 10,
        "J": 11,
        "Q": 12,
        "K": 13,
        "A": 14,
    }
    value_list = []
    for i in hand:
        if i in values:
            value_list += [values[i]]
        else:
           value_list += [int(i)]
    
    return value_list


def determine_hand_value(hand, bid):
    value_list = recode_hand(hand)
    freq_dict = {}
    for value in value_list:
        if value in freq_dict:
            freq_dict[value] += 1
        else:
            freq_dict[value] = 1
    freqs = [item for key, item in freq_dict.items()]
    freqs.sort(reverse=True)

    val = 1
    if len(freqs) == 1:
        val = 7
    elif freqs[0] == 4:
        val = 6
    elif freqs[0] == 3 and freqs[1] == 2:
        val = 5
    elif freqs[0] == 3 and freqs[1] == 1:
        val = 4
    elif freqs[0] == 2 and freqs[1] == 2:
        val = 3
    elif freqs[0] == 2 and freqs[1] == 1:
        val = 2
    
    ranking_list = [val]
    ranking_list += [a for a in value_list]
    ranking_list += [bid]
    return ranking_list

all_hands_processed = []
for hand in data:
    cards = hand[0]
    bid = hand[1]
    all_hands_processed += [determine_hand_value(cards, bid)]

def custom_sort(lst):
    return tuple(lst)

sorted_lists = sorted(all_hands_processed, key=custom_sort)

for i, lst in enumerate(sorted_lists, start=1):
    lst += [i]

output1 = 0
for game in sorted_lists:
    output1 += int(game[6]) * game[7]

print(f"Part 1: {output1}")
```

```python
# Part 2

def recode_hand2(hand):
    hand = list(hand)
    values = {
        "T": 10,
        "J": 1,
        "Q": 12,
        "K": 13,
        "A": 14,
    }
    value_list = []
    for i in hand:
        if i in values:
            value_list += [values[i]]
        else:
           value_list += [int(i)]
    
    return value_list

def determine_hand_value2(hand, bid):
    value_list = recode_hand2(hand)
    freq_dict = {}
    jokers = 0
    for value in value_list:
        if value not in freq_dict:
            if value == 1:
                jokers += 1
            else:
                freq_dict[value] = 1
        else:
            freq_dict[value] += 1

    freqs = [item for key, item in freq_dict.items()]
    freqs.sort(reverse=True)
    if jokers == 5:
        freqs = [5]
    else:
        freqs[0] = freqs[0] + jokers

    val = 1
    if freqs[0] == 5:
        val = 7
    elif freqs[0] == 4:
        val = 6
    elif freqs[0] == 3 and freqs[1] == 2:
        val = 5
    elif freqs[0] == 3 and freqs[1] == 1:
        val = 4
    elif freqs[0] == 2 and freqs[1] == 2:
        val = 3
    elif freqs[0] == 2 and freqs[1] == 1:
        val = 2
    
    ranking_list = [val]
    ranking_list += [a for a in value_list]
    ranking_list += [bid]
    return ranking_list

all_hands_processed = []
for hand in data:
    cards = hand[0]
    bid = hand[1]
    all_hands_processed += [determine_hand_value2(cards, bid)]

def custom_sort(lst):
    return tuple(lst)

sorted_lists = sorted(all_hands_processed, key=custom_sort)

for i, lst in enumerate(sorted_lists, start=1):
    lst += [i]

output2 = 0
for game in sorted_lists:
    output2 += int(game[6]) * game[7]

print(f"Part 2: {output2}")
```
